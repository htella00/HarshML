```python
#importing the required libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
#loading the file into a dataframe
df = pd.read_csv("C:\Python stuff\insurance.csv")
```

```python
#converting the categorical variables "sex", "region", "smoker" to binary
df2 = pd.get_dummies(df)

df2
```

```python
df.describe().T
```

```python
#creating two new columns "overweight" and "unhealthy" for better prediction
df2["overweight"]= (df2.bmi >= 30).astype(int)
df2['unhealthy'] = df2.smoker_yes * df2.overweight
```

```python
#defining the dependent and independent variables
x = df2[['age', 'bmi', 'children', 'sex_female', 'sex_male','smoker_no', 'smoker_yes', 'region_northeast', 'region_northwest','region_southeast', 'region_southwest','overweight','unhealthy']]
y = df2[['charges']]
```

```python
#splitting the dataset into train and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
#developing a linear regression model
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
model = LinearRegression()
model.fit(x_train, y_train)
#y_pred = lr.predict(X_test)
print("R squared for Training score: ", model.score(x_train, y_train)) #model.score displays R-Square
print("R squared for Test score: ", model.score(x_test, y_test))
y_pred_lr = model.predict(x_test)
#print("R Squared: ",model.score(y_test,y_pred_lr))
mse_lr = mean_squared_error(y_test,y_pred_lr)
#mae_lr = mean_absolute_error(y_test, y_pred_lr)
print("The mean squared error is: ",mse_lr)
#print("The mean absolute error is: ",mae_lr)
```

```python
#developing a lasso regression
from sklearn.linear_model import Lasso
model = Lasso() 
model.fit(x_train, y_train)
print("R squared for Training score: ", model.score(x_train, y_train)) #model.score displays R-Square
print("R squared for Test score: ", model.score(x_test, y_test))
y_pred_lasso = model.predict(x_test)
mse_lasso = mean_squared_error(y_test,y_pred_lasso)
mae_lasso = mean_absolute_error(y_test, y_pred_lasso)
print("The mean squared error is: ",mse_lasso)
print("The mean absolute error is: ",mae_lasso)
```
